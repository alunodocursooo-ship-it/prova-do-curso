```python
peso = float(input('qual é o seu peso?'))
altura = float(input('qual é a sua altura?'))
imc = peso/altura**2
if imc <18.5:
  print('abaixo do peso')
elif imc <24.9:
  print('normal')
elif imc <29.9:
  print('sobrepeso')
elif imc <34.9:
  print('obesidade')
else:
  print('obesidade grave')
```
